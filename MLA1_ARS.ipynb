```python
# Write the code to display a list of all of the columns in the dataset.
print("All columns of df:")
print(list(df.columns))
```

```python
# Rename the columns according to this list {date_time, temp_c, dew_point, relative_humidity, wind_speed, visibility, pressure, and weather}.
df.columns = ['date_time', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']
df.head()
```

```python
# Most of the data in the dataset are numerical, write the code to get some standard statistics for each column (e.g. count, mean, etc.)
df.describe()
```

```python
# Write the code to add a column named temp_f to the dataset that contains the temperature from temp_c represented in Fahrenheit.
df['temp_f'] = df['temp_c'] * (9/5) + 32
df.head()
```

```python
# Write the code to rearrange the columns in the dataset so temp_f appears immediately before temp_c.
df = df[['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather']]
df.head()
```

```python
# Write the code to display the data types of all the columns in the dataset.
for c in df.columns:
  print(f'{c}: {type(df.iat[0, df.columns.get_loc(c)])}')
```

```python
# The date_time column is not actually considered a date by Pandas.
# Write the code to convert this column to a date/time column and to verify the conversion was successful.
df['date_time'] = pd.to_datetime(df['date_time'])
print(df['date_time'].dtype)
```

```python
# Write the code to get the average temperature (in F) for each month and write it into an object called monthly_avg.
# This object should be a Pandas Series which is similar to a Pandas Dataframe except it only has one column.
monthly_avg = df.groupby(df['date_time'].dt.to_period('M'))['temp_f'].mean()
monthly_avg
```

```python
# Import graphing utilities
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write code to generate a scatterplot that compares the temperature against the relative humidity. The size of your plot should be 20 x 10.
plt.figure(figsize=(20,10))
sns.scatterplot(x=df["temp_f"], y=df["relative_humidity"])
```

```python
# Write the code to generate a line chart that shows the average temperature per month. The size of your plot should 20 x 10.
plt.figure(figsize=(20,10))
sns.lineplot(x=df['date_time'].dt.strftime('%Y-%m'), y=df["temp_f"])
```

```python
# Write the code to generate a box plot. The size of your plot should 20 x 10.
plt.figure(figsize=(20,10))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.tight_layout()
```
